```python
#loading the dataset
df=pd.read_csv('heart_disease.csv')
df.head()
```

```python
#show number of people with heart diseases
#column target
df.target.value_counts()
```

```python
#show male and female
count_female=len(df[df.sex==0])
count_male=len(df[df.sex==1])
print(count_female)
print(count_male)
```

```python
#display output in percentage
per_female=(count_female/(count_female+count_male))*100
per_female
```

```python
per_male=(count_male/(count_female+count_male))*100
per_male
```

```python
df.groupby('target').mean()
```

```python
#barplot showing heart disease freq of age
pd.crosstab(df.age,df.target).plot(kind='bar',figsize=(50,50))
plt.title('heart disase frequency for Ages')
plt.xlabel('Age')
plt.ylabel('Freq')
plt.show()
```

```python
#Example for testing and evaluation of data

\
from sklearn.model_selection import train_test_split

# Assuming you have a DataFrame called df with columns 'age', 'sex', and 'target'
features = df[['age', 'sex']].values
target = df['target'].values

# Split the data into training and temporary sets (80% training, 20% temporary)
features_train_temp, features_test, target_train_temp, target_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Split the temporary set into training and validation sets (60% training, 20% validation)
features_train, features_val, target_train, target_val = train_test_split(features_train_temp, target_train_temp, test_size=0.25, random_state=42)

# Print the shapes of the resulting sets
print("Training set shape:", features_train.shape)
print("Validation set shape:", features_val.shape)
print("Testing set shape:", features_test.shape)

```
